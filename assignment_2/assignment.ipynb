# Part One

```python
def f(x=0):
    """This function calculates the value of x^3-x^2-1
    Parameters: x ------ any value
    
    Returns: x^3 - x^2 - 1 ---- A function
    """
    return (x**3)-(x**2)-(1)
```

```python
f(2)
```

```python
def df(x=0):
    """This function calculates the value of the derivative of f at point x otherwise, 3x^2 - 2x
    Parameters: x ------ any value
    
    Returns: 3x^2 - 2x --------- The derivative of function f
    """
    return (3*(x**2))-(2*x)
```

```python
df(2)
```

# Part Two

```python
def newton(f, df, x0, epsilon=1e-6, max_iter=30):
    """This function performs a Newton Iteration of the function f, which finds the root Xn, such that f(Xn)=0.
        Parameters
        ----------
                f : function
                Any function
    
                df : function 
                The derivative of the function
                Prerequisite = df(x) != 0
                
                x0 : float
                Any value, which is used as a starting value
                
                epsilon : float
                A value close to 0, such that f(Xn) must be over `epsilon` in order to continue iterating 
                
                max_iter : int
                The maximum amount of iterations
        Returns
        -------
                success_msg : str
                    Human readable message stating `Found root in n iterations`, with n being an int value.
                success_val : int
                    Xn, number of iterations
                
                err_msg : str 
                    Human readable error message stating `Iteration failed` .
                err_val : None
                    None on success

    """
    
    n = 1
    
    x = x0
    
    while n <= max_iter:
         
        x = x - (f(x)/df(x))
        
        if abs(f(x)) < epsilon:
            print ('Found root in %d iterations.' % n)
            return x
        else:  
            n = n + 1
        
    if n > max_iter:
        print('Iteration failed'), 
        return None
    
```

# Part Three

```python
newton(f, df, 1, epsilon=1e-6, max_iter=30)
```
